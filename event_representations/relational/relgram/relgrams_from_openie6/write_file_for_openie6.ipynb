```python
import dill as pickle
from collections import defaultdict
import nltk
```

```python
# Load articles_df.
df_path = '/mount/studenten/arbeitsdaten-studenten1/shencg/BA/V2/data_preprocessing'
articles_df = pickle.load(open(df_path + '/articles_df.p', 'rb'))
articles_df
```

```python
fid2goalsent = dict()
for fid, goal in zip(articles_df['file_id'], articles_df['goal']):
    if fid not in fid2goalsent:
        goal = ["People"] + nltk.word_tokenize(goal)[2:]
        goal = ' '.join(goal)
        fid2goalsent[fid] = goal
print(len(fid2goalsent))
fid2goalsent
```

```python
with open("fid2goalsent.txt", "w") as file:
    for fid, goal in fid2goalsent.items():
        file.write("fid:" + fid + "\n" + goal + ".\n\n")
```
